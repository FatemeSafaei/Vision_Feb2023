```python
import numpy as np
import cv2
```

```python
img = cv2.imread('/content/ball.jpg')
```

```python
img2 = img.reshape((-1, 3))
```

```python
img2 = np.float32(img2)
```

```python
criteria =(cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)

com,labels, centers =  cv2.kmeans(img2, 4,None, criteria, 10, cv2.KMEANS_PP_CENTERS)
```

```python
centers =  np.uint8(centers)
colorized_pixs = centers[labels.flatten()]
```
